```python
path_match = '/content/drive/MyDrive/Colab Notebooks/Cricket projects/t20_json_files/t20_wc_match_results.json'
```

```python
# Open the file for reading
with open(path_match, 'r') as f:
    # Load the JSON data from the file
    data = json.load(f)
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']

  match_ids_dict[key1] = row["match_id"]
  match_ids_dict[key2] = row["match_id"]

match_ids_dict
```

Processing Batting Summary

```python
path_bat = '/content/drive/MyDrive/Colab Notebooks/Cricket projects/t20_json_files/t20_wc_batting_summary.json'
```

```python
# Open the file for reading
with open(path_bat, 'r') as b:
    # Load the JSON data from the file
    data = json.load(b)

data
```

```python
all_records = []
for rec in data:
      all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.head(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†', ''))
df_batting.head(11)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)

df_batting.head(25)
```

```python
# Create the directory if it doesn't exist
!mkdir t20_csv_files
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index = False)
```
